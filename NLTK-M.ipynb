# NLTK

```python
import nltk
import networkx as nx
import re
from nltk.tag import UnigramTagger
from nltk.tag import BigramTagger
```

```python
with open('Data_Sample.txt', 'r', encoding="utf8") as myfile:
    data=myfile.read().replace('\n', '')

```

```python
def ie_preprocess(document):
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences
```

```python
sentences = ie_preprocess(data)
sentences
```

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
for s in sentences:
    result = cp.parse(s)
    print(result)
```

```python
neResult = []
for s in sentences:
    neResult.append(nltk.ne_chunk(s,binary=False))
    nltk.ne_chunk(s,binary=False).draw()
```

```python
is_noun = lambda pos: pos[:2] == 'NN'

nouns = [word for (word, pos) in nltk.pos_tag(nltk.word_tokenize(data)) if is_noun(pos)] 
nouns
```

```python
tagger = UnigramTagger(train)
tagger.evaluate(test)
```

```python

IN = re.compile(r'.*\bin\b(?!\b.+ing)')
for rel in nltk.sem.extract_rels('DT', 'NN', sentences, pattern = IN):
    print(nltk.sem.rtuple(rel))
```
